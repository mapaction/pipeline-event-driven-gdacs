```python
import os
import sys
import pandas as pd


sys.path.append(os.path.abspath(os.path.join(os.getcwd(), "../../../util")))


from data_exploration import (
    check_database,
    fetch_data_from_db,
    fetch_exposed_countries,
)
```

```python
db_path = os.path.abspath(os.path.join(os.getcwd(), "../../data/gdacs_events.db"))

if check_database(db_path):

    df = fetch_data_from_db(db_path)
    if df is not None:

        pd.set_option("display.max_columns", None)
        pd.set_option("display.max_rows", None)
        pd.set_option("display.width", None)
        pd.set_option("display.colheader_justify", "center")

        display(df)
    else:
        print("Failed to fetch data from the database.")

    df_exposed_countries = fetch_exposed_countries(db_path)

    pd.set_option("display.max_rows", None)
    pd.set_option("display.width", None)

    display(df_exposed_countries)
else:
    print("Database file not found.")
```
